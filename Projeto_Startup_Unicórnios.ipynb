# **Análise de Startups Unicórnios com Python**

O projeto consiste em uma análise de startups unicórnios, empresas iniciantes com valor de mercado superior a 1 bilhão de dólares, utilizando a linguagem de programação Python. Os dados para essa análise serão obtidos a partir de uma base de dados pública disponibilizada pelo Kaddle.

**Objetivo:**
O objetivo principal deste projeto é realizar uma análise abrangente das startups que atingiram o status de unicórnio, explorando diversos aspectos como setor de atuação, financiamento, localização geográfica, entre outros. A análise visa fornecer insights e padrões que possam ser úteis para investidores, empreendedores e pesquisadores interessados no ecossistema de startups de alto valor.

**Passo 1:** Bibliotecas Python: Pandas, NumPy, Matplotlib, Seaborn (para manipulação, análise e visualização de dados)

```python
#01: Importando as bibliotecas

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

**Passo 2:** Coleta de Dados: Importação dos dados da base de dados pública do Kaddle para análise no ambiente Python.

```python
#02: Importação das base e Leitura dos dados

Base_Dados = pd.read_csv('unicorns till sep 2022.csv')

#03: Verificar dimensão da base

Base_Dados.shape
```

**Passo 3:** Limpeza e Pré-processamento de Dados: Tratamento de dados ausentes, remoção de duplicatas e transformação de dados.

```python
#04: Lendo a base de dados

Base_Dados.head()
```

```python
#05: Verificando e renomeando as colunas

Base_Dados.rename(columns={
    'Company': 'Empresa',
    'Valuation ($B)': 'Valor ($)',
    'Date Joined': 'Data de adesão',
    'Country': 'País',
    'City': 'Cidade',
    'Industry': 'Setor',
    'Investors': 'Investidores'
}, inplace=True)
```

```python
#06: Verificar tipo de dados da base

Base_Dados.info()
```

```python
#07:Verificando os campos null

Base_Dados.isnull().sum()
```

```python
#08: Visão gráfica com heatmap

plt.figure( figsize=(15,6))
plt.title('Análise dos campos nulos')
sns.heatmap( Base_Dados.isnull(),cbar=False);
```

```python
#09 - Verificando os campos unicos
Base_Dados.nunique()
```

```python
#10 Campos unicos dentro de uma coluna

Base_Dados['Setor'].unique()
```

```python
#11 - Valores Unicos - Contando quantos registros tem na base

Base_Dados['Setor'].value_counts()
```

```python
#12 - Valores unicos por porcentagem

Base_Dados['Setor'].value_counts( normalize=True)
```

**Passo 4:** Análise Exploratória de Dados: Exploração das características das startups unicórnios, identificação de tendências e padrões.

**Passo 5:** Visualização de Dados: Utilização de gráficos e visualizações para comunicar insights.

**Insights:**
A análise revela que o setor com o maior número de unicórnios é o das fintechs, seguido de perto pelos setores de internet e software. Essa tendência destaca a crescente importância da tecnologia financeira e da inovação digital no panorama das startups de alto valor. A predominância desses setores sugere oportunidades significativas para investidores e empreendedores que buscam áreas com alto potencial de crescimento e impacto no mercado global.

**Análise dos Tops 5 Países dos Unicórnios:**
A análise dos países revela uma distribuição interessante dos unicórnios pelo mundo. Observa-se uma concentração significativa em países como Estados Unidos, China e Índia, refletindo a liderança dessas nações no cenário global de inovação e empreendedorismo. Essa análise ressalta a importância da colaboração internacional e da identificação de oportunidades de investimento em diversos mercados ao redor do mundo.

```python
#13 - Analise dos Paises dos unicornios

Analise = round (Base_Dados['País'].value_counts( normalize=True)*100,1)
```

```python
#14 - Visão gráfica com pie

plt.figure( figsize=(15,6))
plt.title('Analise dos Paises Geradores de Unicornios - Top 5')
plt.pie(
    Analise.head(5),
    labels = Analise.index[0:5],
    shadow=True,
    startangle=90,
    autopct='%1.1f%%'
);
```

```python
#15 - Convertendo o campo em tipo 'data'

Base_Dados['Data de adesão'] = pd.to_datetime( Base_Dados ['Data de adesão'])

Base_Dados ['Data de adesão'].head()
```

```python
#16 - Extraindo ano e mês

Base_Dados['Mes'] = pd.DatetimeIndex( Base_Dados['Data de adesão']).month
Base_Dados['Ano'] = pd.DatetimeIndex( Base_Dados['Data de adesão']).year

Base_Dados.head()
```

```python
#17 - Criando uma tabela Analitica

Analise_Agrupada = Base_Dados.groupby( by=['País', 'Ano', 'Mes','Empresa']).count()['Setor'].reset_index()

Analise_Agrupada.head()
```

**Unicórnios no Brasil:**
A análise dos unicórnios no Brasil revela um cenário vibrante e em constante evolução no ecossistema de startups do país. A presença desses unicórnios não apenas impulsiona a economia local, mas também atrai atenção internacional para o potencial do mercado brasileiro.

```python
#18 - Unicornios no Brasil

Analise_Agrupada.loc[
Analise_Agrupada['País'] == 'Brazil'
]
```

```python
#20 - Tranformando a coluna valor e convertendo em valor inteiro

Base_Dados['Valor ($)'] = pd.to_numeric( Base_Dados['Valor ($)'].apply( lambda Linha: Linha.replace('$', '')))

Base_Dados.head()
```

**Análise de valor por País:**

Para realizar uma análise abrangente da distribuição de valor das startups unicórnios por país, foi utilizada uma operação de agrupamento dos dados da base de dados. Primeiramente, os dados foram agrupados pelo campo "País" utilizando o método groupby(). Em seguida, foi aplicada uma operação de soma para calcular o valor total das avaliações das startups unicórnios em cada país, utilizando a coluna "Valor ($)" como referência. Isso nos permitiu obter uma visão consolidada do valor total acumulado das startups unicórnios em cada país.

Após a aplicação da operação de soma, os resultados foram reorganizados em um novo DataFrame utilizando o método reset_index() para garantir uma estrutura clara e legível. O novo DataFrame resultante, denominado "Analise_Pais", contém informações resumidas sobre o valor total das startups unicórnios agrupadas por país.

```python
#21 -  Visão por tabela
Analise_Pais = Base_Dados.groupby(by=['País']).sum()['Valor ($)'].reset_index()

Analise_Pais.head()

```

```python
#22 - Analise dos maiores valores dos unicornios por paises

Analise_Valor = Analise_Pais.sort_values('Valor ($)', ascending=False)
Analise_Valor.head()
```

**Análise de Valor por País:**


A análise de valor por país oferece uma visão abrangente da distribuição da avaliação das startups unicórnios em diferentes regiões do mundo. Os Estados Unidos lideram com uma valorização considerável, impulsionada pelo seu ambiente empreendedor altamente desenvolvido e pelos mercados de capital robustos. A China também apresenta uma forte presença em termos de valor, refletindo o crescimento rápido e a escala das empresas chinesas.

Além disso, observa-se que países como Índia, Brasil, Reino Unido e Alemanha têm unicórnios bem avaliados, indicando a diversificação e o potencial de crescimento em várias partes do mundo. Essa análise não apenas destaca a importância de centros tradicionais de inovação, mas também reconhece o surgimento de novos polos de empreendedorismo globalmente. Esses insights são essenciais para investidores e formuladores de políticas que buscam entender as tendências de valorização e identificar oportunidades de investimento em todo o mundo.

```python
#Visão Gráfica com plot

plt.figure(figsize=(15, 6))
plt.plot(Analise_Valor['País'], Analise_Valor['Valor ($)'])
plt.xticks(rotation=45, ha='right');
```
