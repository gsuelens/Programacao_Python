# **Análise da Base de Dados de um Restaurante Utilizando Gráficos do Seaborn**

Neste projeto, realizei uma análise exploratória de uma base de dados de um restaurante usando as poderosas ferramentas de visualização da biblioteca Seaborn em Python. Meu objetivo foi compreender melhor a distribuição e as relações entre as variáveis do conjunto de dados, utilizando gráficos de dispersão, boxplots e outros recursos oferecidos pelo Seaborn.

**Passo 1: Importando as Bibliotecas negrito**

Nesta etapa inicial do projeto, comecei importando as bibliotecas fundamentais necessárias para realizar a análise exploratória dos dados. As bibliotecas importadas incluem:

```python
#Passo_1: Importando as bibliotecas

import numpy as np
import pandas as pd
import seaborn as sns
```

**Passo 2: Definindo Tema de visualização e Carregando a Base de Dados 'tips' do Seaborn.**

 Nesta etapa do projeto, estabeleci um tema visual consistente para as visualizações que serão criadas ao longo da análise exploratória de dados e carreguei a base de dados 'tips' embutida no Seaborn, que contém informações sobre gorjetas em um restaurante.

```python
#Passo_2: Dinifindo tema para a visualização e carregando a base de dados 'tips' do seanborn

sns.set_theme(style ='darkgrid')

Base_Dados = sns.load_dataset('tips')
```

**Passo 3: Verificando a Base de Dados**

Nesta etapa da análise exploratória de dados, dei início à investigação da base de dados 'tips', verificando suas primeiras linhas usando o método head(). Essa prática é fundamental para obter uma compreensão inicial dos dados e identificar possíveis padrões, tendências ou problemas nos dados.

```python
#Passo_3: Verificar base de dados

Base_Dados.head()
```

**Passo 4: Renomeando as Colunas da Base de Dados**

Nesta etapa da análise de dados, realizei a renomeação das colunas da base de dados 'tips' para torná-las mais descritivas e de fácil compreensão. A renomeação das colunas é uma prática importante para garantir a clareza e a consistência dos dados, facilitando a interpretação e a análise subsequentes.
Utilizei o método rename() do Pandas para renomear as colunas conforme a seguinte correspondência:

```python
#Passo_4: Renomeando as colunas

Base_Dados.rename( columns={
    'total_bill': 'Total Conta',
    'tip': 'Gorjeta',
    'sex':'sexo',
    'smoker':'Fumante',
    'day': 'Dia',
    'time':'Periodo',
    'size':'Pessoas na Mesa'}, inplace=True)
```

**Passo 5: Criando um Gráfico de Dispersão com relplot**

Nesta etapa da análise exploratória de dados, utilizei a função relplot para criar um gráfico de dispersão que representa a relação entre duas variáveis: 'Total Conta' e 'Gorjeta'. Ao plotar 'Total Conta' no eixo x e 'Gorjeta' no eixo y, pude visualizar a relação entre o valor total da conta e a gorjeta deixada pelos clientes do restaurante.

```python
#Pass0_05: Criando um Gráfico de dispersão com relplot
sns.relplot( x ='Total Conta', y='Gorjeta', data=Base_Dados);
```

**Passo 6: Adicionando Parâmetro 'hue' para Colorir os Pontos do Gráfico**


Nesta etapa da análise exploratória de dados, aprimorei o gráfico de dispersão anterior adicionando o parâmetro 'hue' à função relplot do Seaborn. O parâmetro 'hue' é uma ferramenta que permite colorir os pontos do gráfico com base em uma variável categórica, neste caso, 'sexo'. Ao adicionar 'sexo' como o parâmetro 'hue', pude visualizar a distribuição dos pontos no gráfico de dispersão diferenciados por sexo. Essa abordagem de coloração dos pontos facilitou a identificação de possíveis diferenças ou padrões na relação entre 'Total Conta' e 'Gorjeta' entre os diferentes sexos dos clientes do restaurante.

```python
#Passo_06: Adicionando o parametro hue como classe para colorir os pontos do gráfico no gráfico relplot

sns.relplot( x='Total Conta', y='Gorjeta', data=Base_Dados, hue='sexo' );
```

**Passo 7: Criando um Gráfico de Linha com Dois Eixos com Parâmetro 'kind'**

Nesta etapa da análise, utilizei a função relplot do Seaborn para criar um gráfico de linha com dois eixos, especificando o parâmetro 'kind' como 'line'. Esse parâmetro permite que o Seaborn gere automaticamente um gráfico de linha para visualizar a relação entre as variáveis 'Total Conta' e 'Gorjeta'. Ao criar um gráfico de linha, pude observar a tendência geral da relação entre o valor total da conta e a gorjeta deixada pelos clientes do restaurante.

```python
#Passo_07: Criando gráfico de linha com 2 eixo com parametro kind

sns.relplot( x='Total Conta', y='Gorjeta', data=Base_Dados, kind='line' );
```

**Passo 8: Criando um Gráfico de Linha com Dois Eixos e Parâmetro 'hue'**

Utilizei novamente a função relplot do Seaborn para criar um gráfico de linha com dois eixos. Desta vez, adicionei o parâmetro 'hue' para diferenciar as linhas com base na variável categórica 'Fumante'. Ao adicionar 'Fumante' como o parâmetro 'hue', pude visualizar como a relação entre 'Total Conta' e 'Gorjeta' varia entre fumantes e não fumantes.

```python
#Passo_08: Criando gráfico de linha com 2 eixo com parametro kind e hue

sns.relplot( x='Total Conta', y='Gorjeta', data=Base_Dados,
            kind='line', hue='Fumante' );
```

**Passo 9: Criando um Gráfico de Histograma com histplot**

Utilizei a função histplot do Seaborn para criar um gráfico de histograma. O objetivo foi examinar a distribuição dos valores da variável 'Total Conta' na base de dados. Ao criar o histograma para 'Total Conta', pude identificar padrões na distribuição dos valores de conta no restaurante. A altura de cada barra no histograma representa a frequência dos valores dentro de intervalos específicos, permitindo uma rápida compreensão da dispersão dos dados. Isso é crucial para identificar possíveis tendências, outliers ou peculiaridades na distribuição dos valores da variável 'Total Conta'.

```python
#Passo_9: Gráfico histplot

sns.histplot( data=Base_Dados, x='Total Conta' )
```

**Passo 10: Criando um Histograma com parâmetro 'hue'**

Utilizei novamente a função histplot do Seaborn para criar um histograma. Desta vez, adicionei o parâmetro 'hue' para diferenciar as distribuições com base na variável categórica 'Fumante'. Ao adicionar 'Fumante' como o parâmetro 'hue', pude visualizar como a distribuição dos valores da variável 'Total Conta' varia entre fumantes e não fumantes. Isso proporcionou uma compreensão mais detalhada das diferenças na distribuição das contas entre os grupos, permitindo uma análise mais refinada das relações entre as variáveis.

```python
#Passo_10: Gráfico histplot

sns.histplot( data=Base_Dados, x='Total Conta', hue='Fumante' );
```

**Passo 11: Criando um Gráfico de Barra com Barplot**

Utilizei a função barplot do Seaborn para criar um gráfico de barras. O objetivo foi visualizar a relação entre as variáveis 'sexo', 'Total Conta' e 'Fumante', utilizando barras para representar a média do valor total da conta para diferentes categorias de sexo e se são fumantes ou não. No gráfico de barras, o eixo x representa as categorias de 'sexo', enquanto o eixo y representa a média do valor total da conta. As barras são coloridas de acordo com a variável categórica 'Fumante', permitindo uma comparação visual entre os grupos.

```python
#Passo_11: Gráfico Barplot

sns.barplot( data=Base_Dados, x='sexo', y='Total Conta', hue='Fumante');
```

**Passo 12: Criando um Gráfico Pairplot com Todas as Colunas da Base de Dados**

Utilizei a função pairplot para criar um gráfico pairplot. O objetivo foi visualizar as relações entre todas as colunas da base de dados simultaneamente, proporcionando uma visão geral das correlações e distribuições dos dados.
Através do pairplot, pude identificar rapidamente padrões, tendências e relações entre as variáveis da base de dados. Esta visualização é particularmente útil para entender a estrutura dos dados, identificar correlações entre as variáveis e detectar possíveis associações ou agrupamentos.

```python
#Passo_12: Gráfico pairplot: Criando gráfico com todas as colunas da base de dados

sns.pairplot ( Base_Dados);
```

**Passo 12: Análise com Gráfico Pairplot Separando por Sexo**

Utilizei a função pairplot para criar um gráfico pairplot, mas desta vez, com a separação das observações por sexo. O objetivo foi visualizar as relações entre todas as colunas da base de dados, levando em consideração a variável 'sexo', proporcionando uma visão detalhada das correlações e distribuições dos dados para cada categoria de sexo. Através desta análise, pude identificar padrões, tendências e relações específicas para cada categoria de sexo, permitindo uma compreensão mais profunda das características e comportamentos distintos entre homens e mulheres na base de dados.

```python
#Passo_12: Gráfico pairplot: Análise de todas as colunas da base separando pelo sexo

sns.pairplot ( Base_Dados, hue='sexo');
```

**Passo 13: Análise com Gráfico Pairplot com Gráfico de Área**

Utilizei novamente a função pairplot do Seaborn para criar um gráfico pairplot, mas desta vez com o tipo de gráfico definido como 'kde', que representa um gráfico de densidade de kernel (kernel density estimate). Através deste gráfico pairplot com gráfico de área, pude explorar as relações entre todas as colunas da base de dados de forma mais detalhada, observando as densidades de probabilidade conjunta entre pares de variáveis. Isso proporcionou insights valiosos sobre padrões, tendências e associações nos dados, ajudando a identificar possíveis correlações e estruturas subjacentes.

```python
#Passo_13: Gráfico pairplot: Análise de todas as colunas da base com gráfico de area

sns.pairplot ( Base_Dados, kind= 'kde');
```

**Passo 14: Análise do Total das Contas por Dia com Gráfico Boxplot.**

Utilizei a função boxplot para criar um gráfico e analisar a distribuição do total das contas por dia da semana. Ao plotar o total das contas no eixo y e os dias da semana no eixo x, pude visualizar como a distribuição das contas varia ao longo da semana. Isso proporcionou insights valiosos sobre possíveis diferenças nos padrões de consumo dos clientes do restaurante em diferentes dias da semana.

```python
#Passo_14: Análise do total da contas nos dia com a visualiação do  gráfico boxplot

sns.boxplot ( data=Base_Dados, x='Dia', y='Total Conta');
```

**Passo 15: Análise do Total das Contas por Dia e Sexo com Gráfico Boxplot**

Utilizei novamente a função boxplot do Seaborn para criar um gráfico boxplot e analisar a distribuição do total das contas por dia da semana, diferenciando por sexo. Essa abordagem permitiu uma análise mais detalhada das diferenças nos padrões de consumo entre homens e mulheres em diferentes dias da semana. Pude identificar possíveis variações na distribuição do total das contas e entender melhor como essas variações podem estar relacionadas ao sexo dos clientes.

```python
#Passo_15: Análise do total da contas nos dias e pelo sexo com a visualiação do  gráfico boxplot

sns.boxplot ( data=Base_Dados, x='Dia', y='Total Conta', hue='sexo');
```
