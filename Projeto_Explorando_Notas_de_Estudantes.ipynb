# **Projeto: Explorando Notas de Estudantes**

O projeto é uma iniciativa destinada a aprimorar minhas habilidades em análise de dados e visualização utilizando Python. Para este projeto, utilizei uma base de dados pública disponível no Kaggle, contendo informações sobre notas de estudantes.

Além disso, pretendo praticar a utilização de diversas bibliotecas em Python, como Pandas, Matplotlib e Seaborn, para manipulação de dados e criação de visualizações gráficas.

```python
#01 - Importando as libs

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import warnings #usando warnings para ignorar alguns avisos
warnings.filterwarnings('ignore')
```

**Aquisição de Dados:** Baixei a base de dados de notas de estudantes disponível no Kaggle.

```python
#02 - Importando a base de dados

Base_Dados = pd.read_csv('StudentsPerformance.csv')

#03 - Verificando a dimensão da base

Base_Dados.shape
```

```python
#04 - Lendo a base de Dados

Base_Dados.head()
```

**Limpeza e Preparação de Dados:**
Realizei a limpeza dos dados, tratando valores ausentes, eliminando duplicatas e ajustando o formato dos dados, se necessário. Também explorei as características dos dados para entender sua estrutura e distribuição.

```python
#05 - Verificando os campos nulos

Nulos = Base_Dados.isnull()

plt.figure( figsize=(16,5))
plt.title('Análise Campos Nulos')
sns.heatmap(Nulos, cbar=False);
```

```python
#06 - Verificando os campos unicos

Base_Dados.nunique()
```

**Análise de Correlações:** Explorei correlações entre diferentes variáveis, como notas em diferentes disciplinas, nível de escolaridade dos pais e desempenho dos estudantes em testes padronizados.

**Identificação de Insights: **Com base na análise realizada, identifiquei insights interessantes e conclusões sobre o desempenho dos estudantes e possíveis fatores que podem influenciar suas notas.

```python
sns.boxplot( data=Base_Dados, x='math score', y='gender')
```

```python
sns.boxplot( data=Base_Dados, x='reading score', y='gender')
```

```python
sns.boxplot( data=Base_Dados, x='writing score', y='gender')
```

```python
Base_Dados.groupby( by=['gender']).describe()['math score'].reset_index()
```

```python
sns.pairplot( Base_Dados) #com pairplot fezemos gráficos com as variáveis númericas
```

```python
sns.boxplot( data=Base_Dados, x='math score', y='race/ethnicity')
```

```python
sns.boxplot( data=Base_Dados, x='math score', y='parental level of education')
```

```python
Base_Dados.groupby( by=['parental level of education']).describe()['math score'].reset_index()
```

```python
sns.boxplot( data=Base_Dados, x='math score', y='test preparation course')
```

Analisando que a pessoa que se preparou teve uma perfomace melhor

```python
Base_Dados.groupby( by=['test preparation course']).describe()['math score'].reset_index()
```

```python
sns.scatterplot( data=Base_Dados, x='math score', y='writing score')
```

**Resultados:**

Após a análise dos dados, encontramos que a média das notas das mulheres é de 63, enquanto a dos homens é de 68. No entanto, uma observação crucial é que, apesar da média mais alta para os homens, a nota mais alta entre eles é de 27, enquanto uma das mulheres obteve nota zero. Isso indica uma variação significativa nos resultados individuais dentro de cada grupo.

Além disso, ao observar medidas de posição como o primeiro quartil e a mediana, podemos identificar ainda mais nuances nos dados. Essas medidas são importantes para compreender a distribuição das notas e verificar se existem diferenças substanciais entre homens e mulheres em termos de desempenho acadêmico.
